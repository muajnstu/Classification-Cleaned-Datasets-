```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from category_encoders import TargetEncoder
from sklearn.impute import SimpleImputer
```

```python
# load the data
df = pd.read_csv('https://raw.githubusercontent.com/muajnstu/Implementation-of-Active-Learning-Method-in-Regression-Model/refs/heads/main/Preprocessed_dataset_3_.csv')
```

# Basic EDA

```python
df.drop(['applicant_id'], axis=1, inplace=True)
```

```python
df.info()
```

```python
print(df.isnull().sum())
numerical_cols = df.select_dtypes(include=np.number).columns
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])
print("\nAfter imputation:")
print(df.isnull().sum())
```

```python
#function to convert range strings to their mean
def range_to_mean(range_str):
    try:

        return float(range_str)
    except ValueError:
        if isinstance(range_str, str):
            if '-' in range_str:
                low, high = map(float, range_str.split('-'))
            elif ' to ' in range_str:
                low, high = map(float, range_str.split(' to '))
            else:
                return np.nan
            return (low + high) / 2
        else:
            return np.nan
df['cholesterol_level'] = df['cholesterol_level'].apply(range_to_mean)
```

```python
df
```

```python
for col in df.columns:
    print(f"Unique values in column '{col}': {df[col].unique()}")
```

```python
#Encoding

occupation_map = {'Student': 0,'Business': 1, 'Salried': 2}
df['Occupation'] = df['Occupation'].map(occupation_map)
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Alcohol'] = df['Alcohol'].map({'No': 0, 'Rare': 1, 'Daily': 2})
df['exercise'] = df['exercise'].map({'No': 0, 'Moderate': 1, 'Extreme': 2})
smoking_map = {'never smoked': 0,'formerly smoked': 1,'smokes': 2,'Unknown': 3}
df['smoking_status'] = df['smoking_status'].map(smoking_map)

# Target encode Location
target_encoder = TargetEncoder()
df['Location'] = target_encoder.fit_transform(df['Location'], df['insurance_cost'])
```

```python
coverage_map = {'N': 0,'Y': 1}
df['covered_by_any_other_company'] = df['covered_by_any_other_company'].map(coverage_map)
```
